```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/Tiger_reserves_of_India"
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.content,"html.parser")
soup
```

```python
list_of_table_header=soup.find_all("th")  # to find table headers

list_headers=[]
for i in range(0,len(list_of_table_header)):
    a=list_of_table_header[i].get_text()
    list_headers.append(a)
    

import re
final_headers_list=[]
for i in range(0,len(list_headers)):
    a=re.sub("[\n]","",list_headers[i])
    final_headers_list.append(a)

final_headers_list
```

```python
stop_index=final_headers_list.index("Total area (km2)")
stop_index
```

```python
new_final_headers_list=[]

for i in range(0,stop_index+1):
    a=final_headers_list[i]
    new_final_headers_list.append(a)

new_final_headers_list
```

```python
new_final_headers_list.insert(0,"No.")
new_final_headers_list
```

```python
list_table_data=soup.find_all("td")  # to find table data

data_list=[]
for i in range(0,len(list_table_data)):
    a=re.sub("[\n]","",list_table_data[i].get_text())
    data_list.append(a)

data_list
```

```python
stop_index=data_list.index("1,271.47")
stop_index
```

```python
new_data_list=[]

for i in range(0,stop_index+1):
    a=data_list[i]
    new_data_list.append(a)

new_data_list
```

```python
No=[]
for i in range(1,58):
    No.append(i)

No
```

```python
Name=[]
for i in range(0,len(new_data_list),8):
    a=new_data_list[i]
    Name.append(a)

Name
```

```python
Inclusion=[]
for i in range(1,len(new_data_list),8):
    a=new_data_list[i]
    Inclusion.append(a)

Inclusion
```

```python
Last_Notified=[]
for i in range(2,len(new_data_list),8):
    a=new_data_list[i]
    Last_Notified.append(a)

Last_Notified
```

```python
State=[]
for i in range(3,len(new_data_list),8):
    a=new_data_list[i]
    State.append(a)

State
```

```python
Tiger_Population=[]
for i in range(4,len(new_data_list),8):
    a=new_data_list[i]
    Tiger_Population.append(a)

Tiger_Population
```

```python
Core_Area=[]
for i in range(5,len(new_data_list),8):
    a=new_data_list[i]
    Core_Area.append(a)

Core_Area
```

```python
Buffer_Area=[]
for i in range(6,len(new_data_list),8):
    a=new_data_list[i]
    Buffer_Area.append(a)

Buffer_Area
```

```python
Total_Area=[]
for i in range(7,len(new_data_list),8):
    a=new_data_list[i]
    Total_Area.append(a)

Total_Area
```

```python
list_cols=[No,Name,Inclusion,Last_Notified,State,Tiger_Population,Core_Area,Buffer_Area,Total_Area]

data={}
for i in range(0,len(new_final_headers_list)):
    data[new_final_headers_list[i]]=list_cols[i]

data
```

```python
file=pd.DataFrame(data)
file
```

```python
file.to_csv("Tiger_Reserves_of_India.csv")
```
